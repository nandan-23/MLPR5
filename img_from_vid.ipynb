```python
import cv2
import os

destination_folder = 'frames'
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

vidcap = cv2.VideoCapture('Recording 26Sep.mp4')
frame_rate = 10  # Capture a frame every 10 seconds
count = 0
while True:
    success, image = vidcap.read()
    if not success:
        break
    
    if count % (frame_rate * 30) == 0:  # frame_rate * 30 because video frame rate is typically in frames per second
        cv2.imwrite(destination_folder + "/frame%d.jpg" % (count // (frame_rate * 30)), image)  # Save frame as JPEG file
        print('Saved frame %d' % (count // (frame_rate * 30)))

    count += 1

vidcap.release()
cv2.destroyAllWindows()
```
