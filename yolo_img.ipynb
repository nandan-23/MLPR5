<h1>Process a single image</h1>

```python
import os
import cv2

def process_image(image_path, image_len):
    image = cv2.imread(image_path)
    resized_image = cv2.resize(image, (1920, 1080))

    results = model(resized_image, show=False, save=False)
    result = results[0]

    # Extract bounding boxes for 'person' class
    boxes_with_people = [box for box in result.boxes if result.names[box.cls[0].item()] == 'person' and box.conf[0] > 0.85]

    cords_list = []
    for i, box in enumerate(boxes_with_people):
        cords = [round(x) for x in box.xyxy[0].tolist()]
        cords_list.append(cords)
        
        # Update the output filename format
        output_filename = f'im{image_len}__{i+1}.jpg'

        x1, y1, x2, y2 = cords
        cropped_image = resized_image[y1:y2, x1:x2]
        cv2.imwrite(os.path.join(output_dir, output_filename), cropped_image)
        
    return len(boxes_with_people)

image_directory = 'session2_images'
output_dir = 'session2_clippings'
os.makedirs(output_dir, exist_ok=True)

image_len = 1
```

<h1>Process all images in a directory</h1>

```python
# List all files in the specified directory
for filename in os.listdir(image_directory):
    # Check if the file is an image (you can filter by extensions)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        # Create the full path to the image
        image_path = os.path.join(image_directory, filename)
        # Process the image and update image_len
        process_image(image_path, image_len)
        image_len += 1
```
